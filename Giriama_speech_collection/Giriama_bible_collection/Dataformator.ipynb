```python
import pandas as pd

#read the source file for the split. and create the full path to the audio
#it should contain a column of audio file names and a column for their translations 

source_file="/home/eric-kaz/Desktop/Semester4.2/Project /Sem2 /v2.0.swahili/swahiliData/texts/train.csv"
audio_column_name = "wav"
audio_folder = "/home/eric-kaz/Desktop/Semester4.2/Project /Sem2 /v2.0.swahili/swahiliData/wavs/"

df = pd.read_csv(source_file , sep="\t")

df['file_path'] = audio_folder +df["wav"]

df.to_csv("train_clean_wav.csv",index=False)
```

```python
from datasets import load_dataset, Audio  

#create the dataset in hugging face 
dataset = load_dataset("csv", data_files={"train": "/home/eric-kaz/Desktop/Semester4.2/Project /Sem2 /v2.0.swahili/train_clean_wav.csv"})  

dataset = dataset.cast_column("file_path", Audio())

```
